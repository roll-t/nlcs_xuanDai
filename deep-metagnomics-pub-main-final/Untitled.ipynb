```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Load datasets
data_x = pd.read_csv('/mnt/d/20232024-01-CT252-B2003731_final/deep-metagnomics-pub-main-final/data/Chatelier_gut_obesity_x.csv', sep='\t')
data_y = pd.read_csv('/mnt/d/20232024-01-CT252-B2003731_final/deep-metagnomics-pub-main-final/data/Chatelier_gut_obesity_y.csv')

# Merge the datasets using index
merged_data = pd.concat([data_x, data_y], axis=1)

# Select the column 'x' for GMM modeling
selected_column = 'x'  # Update with the appropriate column name

# Prepare data for GMM
data_for_gmm = merged_data[[selected_column]]

# Training Gaussian Mixture Model
num_components = 3  # Adjust number of components as needed
gmm = GaussianMixture(n_components=num_components)
gmm.fit(data_for_gmm)

# Predictions from GMM
labels = gmm.predict(data_for_gmm)
merged_data['cluster'] = labels

# Visualization
colors = ['blue', 'green', 'cyan']  # Modify color list as needed based on the number of components
for k in range(num_components):
    data = merged_data[merged_data['cluster'] == k]
    plt.scatter(data.index, data[selected_column], c=colors[k], label=f'Cluster {k+1}')
plt.xlabel('Index')
plt.ylabel(selected_column)
plt.legend()
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture

# Load the data
data = pd.read_csv('/mnt/d/20232024-01-CT252-B2003731_final/deep-metagnomics-pub-main-final/data/WT2D_x.csv')

# Assuming the first column is the sample identifier, we will drop it
data = data.drop(data.columns[0], axis=1)

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Apply PCA to reduce the data to two components for visualization
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Perform Gaussian Mixture Model clustering
gmm = GaussianMixture(n_components=3, random_state=42)
clusters = gmm.fit_predict(data_pca)

# Plot the clusters
plt.figure(figsize=(14, 7))
colors = ['green', 'blue', 'black']  # Color for each cluster
for i in range(3):
    plt.scatter(data_pca[clusters == i, 0], data_pca[clusters == i, 1], s=15, c=colors[i], label=f'Cluster {i+1}')

# Plot centroids (in GMM, these are not exactly centroids, but the mean of the components)
centroids = gmm.means_
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='red', marker='x', label='Centroids')

plt.title('PCA-reduced GMM Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file
file_path = '/mnt/d/20232024-01-CT252-B2003731_final/deep-metagnomics-pub-main-final/dataset/cancer.xlsx'
data = pd.read_excel(file_path, index_col=0)

# Chuyển đổi dữ liệu từ chuỗi có dấu phẩy sang kiểu số thực chỉ đối với cột kiểu chuỗi
string_columns = data.select_dtypes(include=['object']).columns
data[string_columns] = data[string_columns].apply(lambda x: x.str.replace(',', '.')).astype(float)

# Xây dựng mô hình Gaussian Mixture Model
gmm = GaussianMixture(n_components=10, reg_covar=1.0)  # Chọn số lượng clusters cần phân
gmm.fit(data)

# Dự đoán nhãn của từng điểm dữ liệu
labels = gmm.predict(data)

# Vẽ biểu đồ phân cụm
plt.figure(figsize=(8, 6))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=labels, cmap='viridis')  # Giả định sử dụng hai cột đầu tiên cho việc hiển thị
plt.colorbar(label='Cluster')
plt.xlabel('Sample ID')
plt.ylabel('Cluster')
plt.title('Phân cụm bằng GMM của Dữ liệu vi sinh vật')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your dataset
data = {
    'GMM': [0.85800, 0.66600],
    'MegaD': [0.83300, 0.70000],
    'MegaR': [0.88460, 0.66830],
    'Popphy': [0.91000, 0.65000],
    'MetAML': [0.87700, 0.66400]
}

# Diseases
diseases = [
    'Cirrhosis', 'T2D'
]

np.random.seed(10)  # Đảm bảo kết quả tái tạo được
data_expanded = {}

for method, values in data.items():
    for value, disease in zip(values, diseases):
        # Tạo phân phối giả định xung quanh mỗi giá trị
        data_expanded[f"{method}_{disease}"] = np.random.normal(loc=value, scale=0.1, size=30)

# Tạo một list dữ liệu cho box plot
data_box = [data_expanded[key] for key in data_expanded]

# Tạo figure và axes
fig, ax = plt.subplots(figsize=(7, 5))

# Vẽ boxplot
bplot = ax.boxplot(data_box, patch_artist=True)

# Màu sắc cho mỗi nhóm 6 boxplots
colors = ['blue', 'red', 'green', 'tan', 'pink']  # Thêm màu để phù hợp với số phương pháp

# Tô màu cho các box, thay đổi màu sau mỗi 6 boxes
for i in range(0, len(data_box), len(diseases)):
    color = colors[i // len(diseases)]
    for j in range(len(diseases)):
        bplot['boxes'][i + j].set_facecolor(color)

# Đặt nhãn cho trục x
ax.set_xticklabels([f"{method}_{disease}" for method in data for disease in diseases], rotation=90)

plt.ylabel('Values')
plt.title('Compare data across methods (GMM, MegaD, MegaR, PopPhy, MetAML) for each disease.')
plt.show()
```
